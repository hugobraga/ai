## Geração de Texto e Parâmetros de Customização

```python
# Gera o texto utilizando a nova interface da API
#client = openai.OpenAI(api_key=api_key)
response = client.chat.completions.create(
    model="gpt-3.5-turbo",  # Atualize para o modelo apropriado (por exemplo, gpt-3.5-turbo)
    messages=[
        {"role": "system", "content": "Você é um assistente que cria histórias."},
        {"role": "user", "content": "Era uma vez"}
    ],
    temperature=0.5,  # Controle de aleatoriedade do texto gerado
    max_completion_tokens=150    # Quantidade máxima de tokens para gerar
)
```

O método openai.OpenAI(api_key=api_key).chat.completion.create() aceita vários parâmetros que permitem controlar a saída do texto gerado. Aqui estão alguns dos parâmetros mais comumente usados:

**model**: O ID do modelo que você deseja usar para gerar o texto. Por exemplo, "text-davinci-003" para o GPT-3 (versão mais recente).

**messages**: array contendo papeis (system, user) e textos iniciais que você deseja que o modelo use como ponto de partida para gerar o texto.

**temperature**: Um valor entre 0 e 1 que controla a aleatoriedade do texto gerado. Valores mais altos resultam em texto mais aleatório, enquanto valores mais baixos resultam em texto mais determinístico.

**max_completion_tokens**: O número máximo de tokens que o modelo deve gerar. Note que o modelo pode gerar menos tokens se chegar a uma conclusão natural mais cedo.

**top_p**: Uma alternativa para o parâmetro de temperatura. Este parâmetro controla a diversidade do texto gerado, ao manter apenas as previsões cuja probabilidade acumulada exceda o valor do parâmetro top_p.

**frequency_penalty**: Penalidade para palavras frequentes. O valor deve estar entre -1,0 e 1,0. Valores mais altos desencorajam as palavras mais comuns.

**presence_penalty**: Penalidade para palavras novas. O valor deve estar entre -1,0 e 1,0. Valores mais altos desencorajam as palavras que ainda não apareceram no texto atual.

**stop**: Uma lista de sequências onde a API deve parar de gerar mais tokens.

**n**: O número de diferentes conclusões a serem geradas a partir do prompt dado.

Estes parâmetros podem alterar significativamente a qualidade e o estilo do texto gerado. Você pode querer experimentar diferentes combinações de parâmetros para ver o que funciona melhor para o seu caso de uso específico. 

Além disso, a disponibilidade dessas opções pode variar dependendo da versão específica do modelo e da API da OpenAI que você está usando. Portanto, é sempre uma boa ideia verificar a documentação oficial para obter as informações mais atualizadas.

## Geração de Texto Através de Tradução de Texto

```python
# Prompt
prompt_text = "Traduza o seguinte texto do inglês para o português: \
'A text is a passage of words that conveys a set of meanings to the person who is reading it.'"
```

```python
# Tradução de texto
# client = openai.OpenAI(api_key=api_key)
response = client.chat.completions.create(
    model="gpt-4o-mini",  # Atualize para o modelo apropriado (por exemplo, gpt-4o-mini)
    messages=[
        {"role": "system", "content": "Você é um assistente que traduz textos."},
        {"role": "user", "content": prompt_text}
    ],
    temperature = 0.5,
    max_completion_tokens = 60,
    n = 2  
)
```

```python
print (response)
```

```python
# Print
for i, choice in enumerate(response.choices):
    print(f"Conclusão {i+1}: {choice.message.content.strip()}")
```

## Geração de Texto e Sequências de Parada

O argumento stop_sequences não é diretamente suportado na função Completion.create(). Em vez disso, você tem que criar uma configuração de tokenização mais personalizada para conseguir o efeito de parar na sequência de tokens desejada.

Implementar tal comportamento exige que você gerencie o processo de conclusão por conta própria, possivelmente gerando uma quantidade menor de texto de cada vez e, em seguida, verificando se a sequência de parada foi atingida antes de gerar mais texto. Isso torna o código um pouco mais complexo. Aqui está um exemplo:

```python
# Prompt e stop sequence
prompt = "Escreva um poema sobre a primavera"
stop_sequence = "."
```

```python
# Resposta da API
# client = openai.OpenAI(api_key=api_key)
response = client.chat.completions.create(
    model="gpt-4o-mini",  # Atualize para o modelo apropriado (por exemplo, gpt-4o-mini)
    messages=[
        {"role": "system", "content": "Você é um assistente que escreve poemas."},
        {"role": "user", "content": prompt}
    ],
    temperature = 0.5,
    max_completion_tokens = 60,  # Gerando menos texto de cada vez
)
```

```python
# Carrega a resposta da API
text = response.choices[0].message.content.strip()
print ("primeira leitura de text:\n " + text)
```

```python
# Loop para continuar gerando até atingir a sequência de parada
while not text.endswith(stop_sequence):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Você é um assistente que escreve poemas."},
            {"role": "user", "content": text}  # Somente o texto gerado da última iteração
        ],
        temperature=0.5,
        max_completion_tokens=60,
    )
    
    # Carregar o novo texto gerado
    new_text = response.choices[0].message.content.strip()
    
    # Concatenar o novo texto ao anterior
    text += new_text

    print("texto após concatenação: \n" + text)

# Exibir o poema completo
print("Poema completo:\n" + text)
```

```python
print(text)
```

Nota: Esteja ciente de que esta abordagem aumentará o número de chamadas de API que você faz e, portanto, pode ter um impacto no custo do uso da API da OpenAI. Além disso, este exemplo não considera a possibilidade de que a sequência de parada possa aparecer no meio de um lote de texto gerado, então pode haver situações em que isso não funcionará como esperado, requerendo adaptação.
